```python
import pandas as pd
import matplotlib.pyplot as plt
import io
import base64
```

```python
df = pd.read_csv("pinterest_pins_output.csv")
```

```python
df["datetime"] = pd.to_datetime(df["created_at"])
```

```python
import numpy as np
```

```python
# Sample data for demonstration (replace with your df)

# Bucket to month
df['month'] = df['datetime'].dt.to_period('M')

# Count per month
monthly_counts = df['month'].value_counts().sort_index()
print(monthly_counts.to_csv())

# Plot
fig, ax = plt.subplots(figsize=(10,6))
monthly_counts.plot(kind='bar', ax=ax)
ax.set_title('Events per Month')
ax.set_xlabel('Month')
ax.set_ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('monthly_histogram.png', dpi=150, bbox_inches='tight')
print('Chart saved as monthly_histogram.png')
```
